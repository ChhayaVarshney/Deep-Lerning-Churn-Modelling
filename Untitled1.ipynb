```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv("Churn_Modelling.csv")
```

```python
data_num=data[['CreditScore','Age','Tenure','Balance','NumOfProducts','EstimatedSalary']]
data_cat=data[['Surname','Geography','Gender','HasCrCard','IsActiveMember','Exited']]
```

```python
for i in data_num.columns:
    sns.histplot(data_num[i])
    plt.show()
```

```python
for i in data_cat.columns:
    sns.countplot(data=data,y=data_cat[i],hue='Exited')
    plt.show()
```

```python
# Feature selection and feature engineering
Geography=pd.get_dummies(data['Geography'],drop_first=True)
Gender=pd.get_dummies(data['Gender'],drop_first=True)
data=data.drop('Geography',axis=1)
data=data.drop('Gender',axis=1)
```

```python
data=pd.concat([data,Geography,Gender],axis=1)
```

```python
data.head()
```

```python
data=data.drop('RowNumber',axis=1)
data=data.drop('CustomerId',axis=1)
data=data.drop('Surname',axis=1)
```

```python
y=data['Exited']
X=data.drop('Exited',axis=1)
```

```python
# Splitting data
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
st_x= StandardScaler()
x_train= st_x.fit_transform(x_train)
x_test= st_x.transform(x_test)
```

```python
# Creating an ANN model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import PReLU,ELU,ReLU
from tensorflow.keras.layers import Dropout
```

```python
classifier=Sequential()
# Input Layer
classifier.add(Dense(units=11,activation='relu'))
# Hidden Layers
classifier.add(Dense(units=7,activation='relu'))
classifier.add(Dense(units=5,activation='relu'))
# Output Layer
classifier.add(Dense(1,activation='sigmoid'))
```

```python
# Trainning ANN
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Early Stopping
import tensorflow as tf
Early_stopping=tf.keras.callbacks.EarlyStopping(monitor="val_loss",min_delta=0.0001,patience=20,verbose=1,mode="auto",baseline=None,restore_best_weights=False,)
```

```python
model_history=classifier.fit(x_train,y_train,validation_split=0.33,batch_size=10,epochs=100,callbacks=Early_stopping)
```

```python
model_history.history.keys()
```

```python
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(['train','test'],loc='upper right')
plt.show()
```

```python
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('Model Loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train','test'],loc='upper right')
plt.show()
```

```python
# Prediction of test result
y_pred=classifier.predict(x_test)
y_pred=(y_pred>0.5)

# make confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
# computing the accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
